```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Misalkan kita memiliki DataFrame df dengan fitur-fitur yang ada
df = pd.read_csv('Data Capstone UMKM  - Full.csv')
df.head(5)
```

```python
# Membuat data sintetis yang berisi kolom data target 'pendanaan' dengan linear regression
X = df[['Total Aset (Juta)', 'Penjualan Rata-rata Per Tahun (Juta)', 'Jumlah Tenaga Kerja',
        'Proyeksi Pertumbuhan (%/tahun)', 'Nilai Aset Jaminan Kredit', 'Jumlah Dokumen Kredit']]

df['pendanaan'] = (0.7 * df['Total Aset (Juta)']) + (1.2 * df['Penjualan Rata-rata Per Tahun (Juta)']) + \
                  (0.1 * df['Jumlah Tenaga Kerja']) + (0.3 * df['Proyeksi Pertumbuhan (%/tahun)']) + \
                  (0.1 * df['Nilai Aset Jaminan Kredit']) + (0.1 * df['Jumlah Dokumen Kredit']) + np.random.randn(len(df)) * 10
```

```python
# Inisialisasi model Regresi Linier
model = LinearRegression()

# Fit model ke dataset yang telah kita buat
model.fit(X, df['pendanaan'])

# Prediksi 'pendanaan' menggunakan fitur-fitur yang ada
predictions = model.predict(X)

# Tampilkan lima baris pertama dari DataFrame untuk memeriksa hasilnya
print(df.head(100))
```
